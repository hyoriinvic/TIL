# 행렬 기준으로 값 가져오기

## 열(column) 기준


*   df[열이름]: Pandas의 Series 형태로 반환
*   df[[열목록]]: 2개 이상의 열을 가져올 대, 리스트 형태로 묶어서 지정
  *   Pandas의 DataFrame 형태로 반환





```
df["상호명"].head()
```

```
df[["상호명", "도로명주소"]].head()
```

## 행(row) 기준


*   df.loc[행]
*   df.loc[행, 열]
*   df.loc[행 이름 혹은 번호]: Pandas의 Series 형태로 반환
*   df.loc[[행 목록]]



```
df.loc[0] #행 0을 가져옴
```

```
df.loc[[0, 1, 2], :] #행 0, 1, 2의 모든 열 조회
```

```
#df.loc[0]["상호명"]
df.loc[0, "상호명"] #행 0의 "상호명"
```

```
df.loc[[0, 1, 2], "상호명"] #행 0, 1, 2의 "상호명"
```

```
df.loc[[0, 1, 2], ["상호명", "도로명주소"]] #행 0, 1, 2의 "상호명", "도로명주소"
```

## iloc 로 슬라이싱해 가져오기


*   df.iloc[:]: 전체 데이터 가져오기
*   df.iloc[행, 열] 순으로 인덱스 번호 지정
*  : 는 전체를 의미
* 시작 인덱스: 끝 인덱스 + 1
* : 의 앞이나 뒤에 인덱스를 쓰지 않으면 처음부터 또는 끝까지를 의미함



```
df.iloc[:5, :5] #위에서 5개 행과 왼쪽에서 5개 열
```

```
df.iloc[-5: , -5:] #끝에서 5개 행과 오른쪽에서 5개 열
```

## 구별 음식점 업종 비교하기

```
df_seoul_food = df[(df["시도명"] == "서울특별시") & (df["상권업종대분류명"] == "음식")].copy()
df_seoul_food
```

```
food_gu = df_seoul_food.groupby(["시군구명", "상권업종중분류명"]) #"시군구명"과 "상권업종중분류명"이 같은 행끼리 그룹화
```

```
food_gu.mean()
```

```
food_gu.count()
```

```
food_gu["상호명"].count() #가계이름의 개수 조회
```

```
food_gu = df_seoul_food.groupby(["시군구명", "상권업종중분류명"])["상호명"].count() #상점 수
```

```
food_gu.unstack()
```
