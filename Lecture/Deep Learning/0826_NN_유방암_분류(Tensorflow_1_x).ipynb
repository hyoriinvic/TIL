```
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```
import numpy as np
```

```
import tensorflow as tf
tf.enable_eager_execution()
```

```
data = datasets.load_breast_cancer() #데이터 로드
```

```
X = data.data #독립변수
X = np.array(X, dtype = "float32") #데이터 타입 변환 
```

```
y = data.target #종속변수(0: 악성, 1: 양성)
y = np.array(y, dtype = "float32")
```

```
#.reshape(): 1차원 배열을 2차원 배열로 변환 
##.reshape(1, -1): 행 형태의 2차원 배열로 변환
##.reshape(-1, 1): 열 형태의 2차원 배열로 변환
y = y.reshape(-1, 1)
y
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) #학습/검증 데이터 분할
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
scaler.fit(X_train) #학습 데이터의 평균, 표준편차 계산
```

```
X_train = scaler.transform(X_train) #학습 데이터 표준화
```

```
#활성함수 선언
def sigmoid(h):
  return 1 / (1 + np.exp(-h))
```

```
initializer = tf.contrib.layers.xavier_initializer() #w, b를 초기화 시키는 객체 생성
```

```
w = tf.Variable(initializer([30, 1])) #w 생성(행: 30, 열: 1)
b = tf.Variable(initializer([1])) #b 생성(행: 1, 열: 1)
```

```
tf.matmul(X_train, w) + b #선형 회귀 계산
```

```
hypothesis = tf.sigmoid(tf.matmul(X_train, w) + b) #X_train 선형 회귀 계산
```

```
optimizer = tf.train.AdamOptimizer(0.001) #w, b를 자동으로 수정하는 Optimizer 객체 생성
```

```
#반복해서 최적의 w, b를 찾음
for step in range(100):
  with tf.GradientTape() as tape:
    #로지스틱 회귀식을 변수 hypothesis에 대입
    hypothesis = tf.sigmoid(tf.matmul(X_train, w) + b)
    #로지스틱 회귀식의 오차 계산식을 변수 cost에 대입
    #Tensorflow에서 자동으로 오차를 가장 적게하는 w, b를 찾음
    cost = -tf.reduce_mean(y_train * tf.log(hypothesis) + (1 - y_train) * tf.log(1 - hypothesis))

    #cost, w, b를 이용해 w의 기울기, b의 기울기를 계산해 grads에 대입
    grads = tape.gradient(cost, [w, b])
  
  #w, b 업데이트
  optimizer.apply_gradients(grads_and_vars = zip(grads, [w, b]))

  if step % 10 == 0:
    print("=" * 50)
    print("step: {}, cost: {}, w: {}, b: {}".format(step, cost.numpy(), w.numpy(), b.numpy()))
    print("=" * 50)
```

```
#Gradient Decent를 통해 계산이 완료된 w
w
```

```
#Gradient Decent를 통해 계산이 완료된 b
b
```

```
X_test = scaler.transform(X_test) #X_train의 평균, 표준편차를 이용해 X_test의 범위를 변경
X_test
```

```
predict = tf.sigmoid(tf.matmul(X_test, w) + b) #예측값
```

```
#tf.cast(조건): 조건을 만족하는 데이터 또는 True를 1로 변환 
predict01 = tf.cast(predict > 0.5, dtype = tf.float32)
```

```
#tf.equal(A, B): A = B 이면 True 리턴
ac01 = tf.equal(predict01, y_test)
```

```
#tf.cast(): True는 1로, False는 0으로 변환
ac02 = tf.cast(ac01, dtype = "float32")
```

```
#tf.reduce_mean(): 변환한 0과 1의 합을 구한 후, 데이터의 개수로 나눔
ac03 = tf.reduce_mean(ac02)
ac03
```

```
#정확도
accuracy = tf.reduce_mean(tf.cast(tf.equal(predict01, y_test), dtype = tf.float32))
print("accuracy: ", accuracy.numpy())
```
