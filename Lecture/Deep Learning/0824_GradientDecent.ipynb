```
import numpy as np
```

Gradient Decent 

```
X = np.array([1, 2, 3], dtype = "float32") #DataType이 지정된 Array 
y = np.array([2, 2.5, 3.5], dtype = "float32")
```

```
#뇌피셜 연산: Hypothesis(가설), 또는 H(x)
w = 2
b = 1
hypothesis = (X * w) + b
```

```
#MSE(Mean Squared Error, 또는 cost) 계산
((3 - 2) ** 2 + (5 - 2.5) ** 2 + (7 - 3.5) ** 2) / 3
```

```
#'MSE = 0'이 되는 w, b 를 찾는 것이 목표(갱신)
```

```
np.dot((hypothesis - y), X)
```

```
w_gred = 1/3 * np.dot((hypothesis - y), X) #Gradient Decent 편미분 공식에 따라 w_gred 계산 (기존 w를 얼마나 변화시켜야 하는가)
w_gred
```

```
w = w - w_gred #w 업데이트
w
```

```
#같은 과정에 따라 b_gred 계산 후, b 업데이트
b_gred = 1/3 * np.sum(hypothesis - y)
b = b - b_gred
b
```

```
#Gradient Decent 반복 실행, 오차제곱합이 최소가 되는 w, b 찾기
for i in range(100):
  hypothesis = X * w + b #예측값을 hypothesis에 대입
  print("hypothesis: ", hypothesis)

  MSE = np.sum((hypothesis - y) ** 2) #cost 계산
  print("MSE: ", MSE)

  w_gred = np.dot((hypothesis - y), X) * 1/3
  print("w_gred: ", w_gred)
  w = w - w_gred
  print("w: ", w)

  b_gred = np.sum(hypothesis - y) * 1/3
  print("b_gred: ", b_gred)
  b = b - b_gred
  print("b: ", b)
```

```
#위의 경우, cost가 무한대로 가는 문제 발생 (본 의도: 'cost = 0'을 만들어야 함)
##w_gred와 b_gred가 너무 큰 값으로 계산되어 cost가 0이 되는 지점을 찾지 못하는 것
##이를 'Obershooting Problem' 이라고 함
```

```
#Learning Rate(학습률)
##Overshooting Problem 해결을 위해 w_gred와 b_gred에 곱하는 1보다 작은 소수

w = 2
b = 1
learning_rate = 0.1 

for i in range(100):
  hypothesis = X * w + b #예측값을 hypothesis에 대입
  print("hypothesis: ", hypothesis)

  MSE = np.sum((hypothesis - y) ** 2) #cost 계산
  print("MSE: ", MSE)

  w_gred = np.dot((hypothesis - y), X) * 1/3
  print("w_gred: ", w_gred)
  w = w - (w_gred * learning_rate)
  print("w: ", w)

  b_gred = np.sum(hypothesis - y) * 1/3
  print("b_gred: ", b_gred)
  b = b - (b_gred * learning_rate)
  print("b: ", b)
  print("*" * 100)
```
