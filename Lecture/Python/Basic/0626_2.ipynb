Package random , 랜덤함수

```
import random #랜덤한 숫자 만들 때 쓰는 패키지

num_lst = [1, 2, 3, 4, 5, 6, 7]
hangul_lst = ["가", "나", "다", "라"]

#random.choice(리스트) - 리스트 내에서 랜덤하게 하나를 복사해 리턴
print(random.choice(num_lst))
print(random.choice(hangul_lst))

#random.randrange(범위) - 범위 내에서 랜덤하게 하나를 리턴
print(random.randrange(1, 100)) #1 이상 100 미만의 랜덤한 숫자 하나
print(random.randrange(1, 3)) #1 이상 3 미만의 랜덤한 숫자 하나
```

문자열 관련 함수

```
#문자열을 다루는 함수

#문자열 앞/뒤 공백 삭제 .strip()
line = "     안녕     "
print(line)
print(line.strip())

#긴 문자열을 잘라서 리스트로 만듦 .split(기준 문자열)
line = "안녕, 반가워, 또 만나 ^^"
print(line)
print(line.split(","))
```

```
line = "길동, 87," #데이터가 부족할 때
print(line.split(","))

name, weight, height = line.split(",")
print(name)
print(weight)
print(height) #데이터가 없기에 빈 칸 출력

if name:
  print("A")

#데이터가 없기에 False - if문 실행 X
#if문의 조건식이 0, ''(빈 문자열), None 일 경우, False로 인식
if height: 
  print("A")
else: 
  print("B")

if not height:
  print("A")
```

```
#확인문제 p.269: filter 함수 활용하기 !!
numbers = list(range(1, 10 + 1))

print("#홀수만 추출하기")
print(list(filter(lambda x: x % 2 == 1, numbers))) #인라인 람다
print()

print("#3 이상, 7 미만 추출하기")
print(list(filter(lambda x: 3 <= x < 7, numbers)))
print()

print("#제곱해서 50 미만 추출하기")
print(list(filter(lambda x: x ** 2 < 50, numbers)))
```

```
#직접 해보는 손코딩 p.261-262
import random

hanguls = list("가나다라마바사아자차카타파하") #list() 함수를 사용한 것

with open("/gdrive/My Drive/Colab Notebooks/info.txt", "w") as file:
  for i in range(1000):
    name = random.choice(hanguls) + random.choice(hanguls)
    weight = random.randrange(40, 100) 
    height = random.randrange(140, 200)

    file.write("{}, {}, {}\n".format(name, weight, height)) #줄바꿈을 위해 \n 을 써줘야 함

with open("/gdrive/My Drive/Colab Notebooks/info.txt", "r") as file:
  for line in file: #file의 내용을 한 줄 씩 읽기
    (name, weight, height) = line.strip().split(",") 
    #.strip() 문자열 앞뒤 공백을 제거하고, .split(",") ,를 기준으로 자름
    #튜플을 활용해 코드를 간소화

    if (not name) or (not weight) or (not height):
      continue
    
    bmi = int(weight) / ((int(height)/100) ** 2)
    bmi = round(bmi, 2) #소수점 처리
    #bmi = round(int(weight) / ((int(height)/100) ** 2), 2) 
    result = ""
    
    if 25 <= bmi:
      result = "과체중"
    elif 18.5 <= bmi:
      result = "정상체중"
    else:
      result = "저체중"

    print('\n'. join([
                      "이름: {} \t"
                      "몸무게: {} \t"
                      "키: {} \t"
                      "BMI: {} \t"
                      "결과: {} \t" 

    ]).format(name, weight, height, bmi, result))
    print()
```

```
#텍스트 한 줄씩 읽기

import random

hanguls = list("가나다라마바사아자차카타파하")

with open("info.txt", "w") as file:
  for i in range(5):
    name = random.choice(hanguls) + random.choice(hanguls)
    weight = random.randrange(40, 100)
    height = random.randrange(140, 200)

    file.write("{}, {}, {}\n".format(name, weight, height))

with open("info.txt", "r") as file:
  for line in file:
    (name, weight, height) = line.strip().split(",")

    if (not name) or (not weight) or (not height):
      continue #다음 줄로 넘어감!_!
    
    bmi = int(weight) / ((int(height)/100) * (int(height)/100))
    result = ""
    
    if 25 <= bmi:
      result = "과체중"
    elif 18.5 <= bmi:
      result = "정상체중"
    else:
      result = "저체중"

    print('\n'. join([
                      "이름: {} "
                      "몸무게: {} "
                      "키: {} "
                      "BMI: {} "
                      "결과: {} " 

    ]).format(name, weight, height, bmi, result))
    print()
```
