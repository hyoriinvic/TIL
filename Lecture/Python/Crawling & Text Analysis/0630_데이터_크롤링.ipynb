```
#urllib.request를 이용한 이미지 다운로드
#다운로드 받은 이미지를 메모리에 저장하지 않고 바로 하드디스크에 저장
import urllib.request
```

```
url = "https://postfiles.pstatic.net/20100615_12/kkchblue_12765686609822pxfs_jpg/%EA%B0%95%EC%95%84%EC%A7%80_kkchblue.jpg?type=w2"
filename = "/gdrive/My Drive/Colab Notebooks/img01.jpg" #파일의 본래 확장자와 같은 확장자명을 써야 함 
urllib.request.urlretrieve(url, filename)
```

```
#urllib.request.uriopen()으로 이미지를 파일에 저장
#다운로드 받은 이미지를 메모리에 저장(변수에)한 후 하드디스크에 저장

mem = urllib.request.urlopen(url).read()
```

```
mem #mem에는 숫자(16진수)가 들어있음
```

```
#문자열을 숫자로 변환
import io
io.BytesIO(mem)
```

```
#화면에 이미지를 출력하는 객체
from PIL import Image
```

```
Image.open(io.BytesIO(mem))
```

```
filename = "/gdrive/My Drive/Colab Notebooks/img02.jpg"

with open(filename, "wb") as f:
  f.write(mem)
```
